```python
# import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
import sqlite3
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

# Call dataset

```python
df = pd.read_csv('sales1.csv')
```

# Columns Handiling

```python
df.columns   
```

```python
df.nunique()      # count Unique value in every column
```

```python
df.columns = df.columns.str.lower()      # for column name lower 
df.columns
```

```python
df = df.rename(columns={'row id':'row_id','order id':'order_id','order date':'order_date',
                        'ship date':'ship_date','ship mode':'ship_mode','customer id':'customer_id',
                        'customer name':'customer_name','postal code':'postal_code',
                        'product id':'product_id','product name':'product_name','sub-category':'sub_category'})
```

# Missing Value Handiling

```python
df.isnull().sum()    #count missing Value
```

```python
df['sales'] = df['sales'].fillna(df['sales'].mode().iloc[0])
df['postal_code'] = df['postal_code'].fillna(df['postal_code'].mode().iloc[0])
df['segment'] = df['segment'].fillna(df['segment'].mode().iloc[0])
df['region'] = df['region'].fillna(df['region'].mode().iloc[0])
```

```python
df.isnull().sum()    #again checking missing Value
```

# Statistics

```python
df.describe()
```

```python
df2 = df.copy()    #for making new dataset
```

```python
le = LabelEncoder()                #For label encoding
df2['row_id'] = le.fit_transform(df2['row_id'])
df2['segment'] = le.fit_transform(df2['segment'])
df2['order_id'] = le.fit_transform(df2['order_id'])
df2['order_date'] = le.fit_transform(df2['order_date'])
df2['ship_date'] = le.fit_transform(df2['ship_date'])
df2['ship_mode'] = le.fit_transform(df2['ship_mode'])
df2['customer_id'] = le.fit_transform(df2['customer_id'])
df2['customer_name'] = le.fit_transform(df2['customer_name'])
df2['country'] = le.fit_transform(df2['country'])
df2['city'] = le.fit_transform(df2['city'])
df2['state'] = le.fit_transform(df2['state'])
df2['postal_code'] = le.fit_transform(df2['postal_code'])
df2['region'] = le.fit_transform(df2['region'])
df2['product_id'] = le.fit_transform(df2['product_id'])
df2['category'] = le.fit_transform(df2['category'])
df2['sub_category'] = le.fit_transform(df2['sub_category'])
df2['product_name'] = le.fit_transform(df2['product_name'])
df2['sales'] = le.fit_transform(df2['sales'])
df2 = df2.drop(columns = ['country'])
```

```python
# Connect to SQLite (in-memory)
conn = sqlite3.connect(":memory:")

# Store 'df' DataFrame in SQLite
df.to_sql("df_table", conn, index=False, if_exists="replace")
```

# Total Sales in Category

```python
query = "SELECT category, SUM(sales) AS total_sales FROM df_table GROUP BY category"
sales_summary = pd.read_sql(query, conn)
plt.figure(figsize =(8,5))
p=sns.barplot(x='category',y='total_sales',data=sales_summary,palette = "bright")
plt.xlabel('state',fontsize=16,fontweight='bold')
plt.ylabel('Total_sales',fontsize=16,fontweight='bold')
plt.title('Total Sales by category',fontsize=20,fontweight='bold')
for container in p.containers:
    p.bar_label(container,fontsize=12,
               bbox = {'boxstyle':'circle','facecolor':'yellow','edgecolor':'red'},fontweight = 'bold')
p.set_facecolor("#f4f4f4")
p.grid(False)
plt.show
```

# Total Sales by Region

```python
query1 = "SELECT region, SUM(sales) AS total_sales FROM df_table GROUP BY region"
sales_summary2 = pd.read_sql(query1, conn)
plt.figure(figsize =(8,5))
ap=sns.barplot(x='region',y='total_sales',data=sales_summary2,palette = "bright")
plt.xlabel('region',fontsize=20,fontweight='bold')
plt.ylabel('Total_sales',fontsize=20,fontweight='bold')
plt.xticks(rotation=45)
for container in ap.containers:
    ap.bar_label(container,fontsize=12,
               bbox = {'boxstyle':'round','facecolor':'yellow','edgecolor':'red'},fontweight = 'bold')
ap.set_facecolor("#f4f4f4")
ap.grid(False)
plt.show
```

# Total_sales by State

```python
query2 = "select state, sum(sales) AS Total_sales from df_table GROUP BY state order by Total_sales desc"
sales_summary3 = pd.read_sql(query2, conn)
plt.figure(figsize =(40,28))
ax=sns.barplot(x='Total_sales',y='state',data=sales_summary3,palette = "bright")
plt.xlabel('Total_sales',fontsize=40,fontweight='bold')
plt.ylabel('State',fontsize=40,fontweight='bold')
plt.title('Total Sales by State',fontweight='bold',fontsize=50)
plt.xticks(rotation=45)
for container in ax.containers:
    ax.bar_label(container,fontsize=20,
               bbox = {'boxstyle':'Rarrow','facecolor':'white','edgecolor':'red'},fontweight = 'bold')
ax.set_facecolor("#f4f4f4")
ax.grid(False)
plt.show()
```

Comment: Here we have seen that california state has been most sell.

# Find out the 10 most product has been sell?

```python
query3="select product_name,sum(sales) as Total_sales from df_table group by product_name order by Total_sales desc limit 10"
sales_summary4= pd.read_sql(query3,conn)
plt.figure(figsize=(30,25))
az = sns.barplot(x= 'Total_sales',y='product_name',data=sales_summary4,palette = "bright")
plt.xlabel('Total_sales',fontsize=40,fontweight='bold')
plt.ylabel('Product Name',fontsize=40,fontweight='bold')
plt.title('Most 10 Product sell',fontweight='bold',fontsize=50)
plt.xticks(rotation=45)
for container in az.containers:
    az.bar_label(container,fontsize=20,
               bbox = {'boxstyle':'circle','facecolor':'white','edgecolor':'red'},fontweight = 'bold')
az.set_facecolor("#f4f4f4")
az.grid(False)
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format ='%d/%m/%Y')  ##Convert the column into datetime object 
df['month'] = df['order_date'].dt.month               ##Extracts the month from the 'order_date' column.
df['Year']  = df['order_date'].dt.year                ##Extracts the year from the 'order_date' column.
df['Day']  = df['order_date'].dt.day                ##Extracts the day from the 'order_date' column.
df['Day'] = df['Day'].astype(int)                 ##Change data type 
```

# Find out total sales in every year

```python
query4 = "select Year,month,sum(sales) as Total_Sales from df_table group by Year,month order by Total_Sales desc"
sales_summary6 = pd.read_sql(query4,conn)
plt.figure(figsize=(12, 6))

plt.figure(figsize=(15, 10))
sns.lineplot(data=sales_summary6, x='month', y='Total_Sales', hue='Year', marker='o')

plt.title('Total Sales per Year by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Display all months on the x-axis
plt.legend(title="Year")
plt.grid(True)
plt.show()
```

Comment:we have seen in the  avobe line plot that sales contribution in 2018 has taken better position than others years.
